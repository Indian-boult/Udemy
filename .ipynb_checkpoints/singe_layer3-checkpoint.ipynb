```python
def scrape_html(html):
    soup = bs.BeautifulSoup(html, 'lxml')
    enrollment = soup.find('div', attrs = {'data-purpose' : 'enrollment'})
    last_update = soup.find('div', attrs = {'class': 'last-update-date'})
    language = soup.find('div', attrs = {'class': 'clp-lead__locale'})
    description = soup.find('div', attrs = {'class' : 'clp-lead__headline'})
    categories = [cat.text.strip("\n") for cat in soup.find_all('a', attrs = {'class': 'topic-menu__link'})]
    
    print categories
    print categories
    print categories
    print enrollment.text.strip(" \n").split(" ")[0]
    print language.text.strip("\n")
    print description.text.strip("\n")
    print last_update.text.strip("\n").split(" ")[-1]
    print
    
    data = [categories[0], categories[1], categories[2], enrollment.text.strip(" \n").split(" ")[0], 
           language.text.strip("\n"), description.text.strip("\n"), last_update.text.strip("\n").split(" ")[-1]]
    
    return data
```

```python
new_cols = ['Category 1', 'Category 2', 'Category 3', 'Enrollment', 'Language', 'Description', 'Last Update']
```

```python
all_files = glob.glob('/Users/thai/Desktop/Kris/out'+ "/*.csv")
```

```python
def pull_info(csv_in, proxies, user_agent):
    headers = {'User-Agent': user_agent}
    out_file = os.path.basename(csv_in)
    df = pd.read_csv(csv_in)
    for i in new_cols:
        df[i] = pd.Series()
        
    try:    
        for idx in range(len(df)):
            req_url = 'http://' + df.iloc[idx]["URL"]
            req = requests.get(url=req_url, proxies=proxies, headers=headers) 
#             html = urlopen(req).read()
            html = req.content
            df_single = scrape_html(html)
            df.loc[idx,new_cols] = df_single
        
        df.to_csv('~/Desktop/Kris/final/'+out_file, index = None, header=True,encoding = 'utf-8')
    except HTTPError:
        print "BLOCKED @ " + str(idx) + " /" + str(len(df))
        agent = ua.random
        headers = {'User-Agent': agent}
        pull_info(csv_in,proxies, agent)

    return df
```
